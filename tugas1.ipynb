```python
# Soal_01.py - Analisis Nilai Mahasiswa (Array)
import matplotlib.pyplot as plt

# 1. Input 10 nilai
nilai = []
for i in range(10):
    while True:
        try:
            n = float(input(f"Nilai {i+1}: "))
            if 0 <= n <= 100:
                nilai.append(n)
                break
            print("Nilai 0-100")
        except:
            print("Angka valid")
```

```python
# 2. Nilai tertinggi & terendah [O(n)]
max_n = max(nilai)  # O(n)
min_n = min(nilai)  # O(n)
print("\nHASIL ANALISIS:")
print(f"Tertinggi: {max_n}")
print(f"Terendah: {min_n}")
```

```python

# 3. Rata-rata [O(n)]
rata = sum(nilai) / 10  # O(n)
print(f"Rata-rata: {rata:.1f}")

# 4. Jumlah lulus [O(n)]
lulus = sum(1 for x in nilai if x >= 60)  # O(n)
print(f"Lulus (>=60): {lulus}/10")
```

```python
# 5 dan 6 . Grafik [O(1)]
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Grafik 1: Tertinggi/Terendah
ax1.bar(['Max', 'Min'], [max_n, min_n], color=['green', 'red'])
ax1.set_ylim(0, 100)
ax1.set_title('Nilai Tertinggi')

# Grafik 2: Kelulusan
ax2.bar(['Lulus', 'Gagal'], [lulus, 10-lulus], color=['green', 'red'])
ax2.set_ylim(0, 10)
ax2.set_title('Kelulusan')

plt.suptitle('Analisis Array Nilai')
plt.tight_layout()
plt.show()
```
